```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder , LabelEncoder
#Colocando os dados em ordem aleatória 
randomdata = (df.sample(n=1030, replace=False, random_state = 42))
from sklearn.model_selection import KFold, cross_val_score, train_test_split
#Treino e teste

#Treino e teste
X = randomdata.drop('resistência', axis = 1)
y = randomdata['resistência']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

#Padronizando dados
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (RandomForestRegressor, GradientBoostingRegressor, 
                              AdaBoostRegressor)

###Regressao Linear
LIR = LinearRegression()
LIR = LIR.fit(X_train, y_train)
pred_lir = LIR.predict(X_test)
print('Regressão Linear:')
print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_lir))
print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_lir)))
print('R2:', r2_score(y_test, pred_lir))
lir0 = np.sqrt(mean_squared_error(y_test, pred_lir))
#Validação cruzada
lir_cv = cross_val_score(estimator = LIR, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')
print("Erro Quadrático Médio usando validação cruzada:",(lir_cv.mean()*(-1)))
print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((lir_cv.mean()*(-1))), "\n")
lir1 = np.sqrt((lir_cv.mean()*(-1)))

#Regressao Polinomial Grau 4
#https://www.kaggle.com/kevinscaria/predict-compressive-strength-of-concrete
from sklearn.preprocessing import PolynomialFeatures
#poly = PolynomialFeatures(degree = 4)
#X_polytr = poly.fit_transform(X_train)
#poly4 = X_polytr.fit(X_train, y_train)
#pred_poly = poly.predict(X_test)
#print('Regressão Polinomial Grau 4:')
#print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_poly))
#print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_poly)))
#poly0 = np.sqrt(mean_squared_error(y_test, pred_poly))
#Validação cruzada
#poly_cv = cross_val_score(estimator = poly, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')
#print("Erro Quadrático Médio usando validação cruzada:",(poly_cv.mean()*(-1)))
#print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((poly_cv.mean()*(-1))), "\n")
#poly1 = np.sqrt((poly_cv.mean()*(-1)))


#Lasso Regression
from sklearn.linear_model import Ridge,Lasso
LASSO = Lasso()
LASSO = LASSO.fit(X_train, y_train)
pred_lasso = LASSO.predict(X_test)
print('Regressão LASSO:')
print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_lasso))
print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_lasso)))
print('R2:', r2_score(y_test, pred_lasso))
lasso0 = np.sqrt(mean_squared_error(y_test, pred_lasso))
#Validação cruzada
lasso_cv = cross_val_score(estimator = LASSO, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')
print("Erro Quadrático Médio usando validação cruzada:",(lasso_cv.mean()*(-1)))
print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((lasso_cv.mean()*(-1))), "\n")
lasso1 = np.sqrt((lasso_cv.mean()*(-1)))

##Regressão de Ridge
ridge  = Ridge()
ridge  = ridge.fit(X_train, y_train)
pred_ridge  = ridge.predict(X_test)
print('Regressão de Rigde :')
print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_ridge))
print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_ridge)))
print('R2:', r2_score(y_test, pred_ridge))
ridge0 = np.sqrt(mean_squared_error(y_test, pred_ridge))
#Validação cruzada
ridge_cv = cross_val_score(estimator = ridge, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')
print("Erro Quadrático Médio usando validação cruzada:",(ridge_cv.mean()*(-1)))
print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((ridge_cv.mean()*(-1))), "\n")
ridge1 = np.sqrt((ridge_cv.mean()*(-1)))

###KNN
KNN = KNeighborsRegressor(n_neighbors=3)
KNN = KNN.fit(X_train, y_train)
pred_knn = KNN.predict(X_test)
print('K vizinhos mais próximos:')
print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_knn))
print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_knn)))
print('R2:', r2_score(y_test, pred_knn))
knn0 = np.sqrt(mean_squared_error(y_test, pred_knn))
#Validação cruzada
knn_cv = cross_val_score(estimator = KNN, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')
print("Erro Quadrático Médio usando validação cruzada:",(knn_cv.mean()*(-1)))
print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((knn_cv.mean()*(-1))), "\n")
knn1 = np.sqrt((knn_cv.mean()*(-1)))

#Ada Boost Regressor
from sklearn.ensemble import AdaBoostRegressor
adb = AdaBoostRegressor()
adb = adb.fit(X_train, y_train)
pred_adb = adb.predict(X_test)
print('Regressão AdaBoost:')
print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_adb))
print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_adb)))
print('R2:', r2_score(y_test, pred_adb))
adb0 = np.sqrt(mean_squared_error(y_test, pred_adb))
#Validação cruzada
adb_cv = cross_val_score(estimator = adb, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')
print("Erro Quadrático Médio usando validação cruzada:",(adb_cv.mean()*(-1)))
print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((adb_cv.mean()*(-1))), "\n")

adb1 = np.sqrt((adb_cv.mean()*(-1)))

###Árvore de decisão
DTR = DecisionTreeRegressor(random_state = 42)
DTR = DTR.fit(X_train, y_train)
pred_dtr = DTR.predict(X_test)
print('Arvores de Decisão:')
print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_dtr))
print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_dtr)))
print('R2:', r2_score(y_test, pred_dtr))
dtr0 = np.sqrt(mean_squared_error(y_test, pred_dtr))
#Validação cruzada
dtr_cv = cross_val_score(estimator = DTR, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')
print("Erro Quadrático Médio usando validação cruzada:",(dtr_cv.mean()*(-1)))
print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((dtr_cv.mean()*(-1))), "\n")
dtr1 = np.sqrt((dtr_cv.mean()*(-1)))

###Random Forest
RFR = RandomForestRegressor(n_estimators=200, random_state = 42)
RFR = RFR.fit(X_train, y_train)
pred_rfr = RFR.predict(X_test)
print('Florestas aleatórias de decisão:')
print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_rfr))
print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_rfr)))
print('R2:', r2_score(y_test, pred_rfr))
rfr0 = np.sqrt(mean_squared_error(y_test, pred_rfr))
#Validação cruzada
rfr_cv = cross_val_score(estimator = RFR, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')
print("Erro Quadrático Médio usando validação cruzada:",(rfr_cv.mean()*(-1)))
print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((rfr_cv.mean()*(-1))), "\n")
rfr1 = np.sqrt((rfr_cv.mean()*(-1)))

#Regressão Gradiente Boosting
from sklearn.ensemble import GradientBoostingRegressor
gbr = GradientBoostingRegressor()
gbr = gbr.fit(X_train, y_train)
pred_gbr = gbr.predict(X_test)
print('Regressão Gradiente Boosting:')
print('Erro Quadrático Médio:', mean_squared_error(y_test, pred_gbr))
print('Raiz do Erro Quadrático Médio:', np.sqrt(mean_squared_error(y_test, pred_gbr)))
print('R2:', r2_score(y_test, pred_gbr))
gbr0 = np.sqrt(mean_squared_error(y_test, pred_gbr))
#Validação cruzada
gbr_cv = cross_val_score(estimator = gbr, X = X_train, y = y_train, cv = 10, scoring='neg_mean_squared_error')

print("Erro Quadrático Médio usando validação cruzada:",(gbr_cv.mean()*(-1)))
print('Raiz do Erro Quadrático Médio usando validação cruzada:', np.sqrt((gbr_cv.mean()*(-1))), "\n")
#print('R2 usando validação cruzada:', r2_score(y,cross_val_predict(gbr_cv,X_train,y_train,cv=10)))

gbr1 = np.sqrt((gbr_cv.mean()*(-1)))


conc0 = {
    'Holdout':[lir0,lasso0,ridge0,knn0,adb0,dtr0,rfr0,gbr0] ,
    'CrossValidation':[lir1,lasso1,ridge1,knn1,adb1,dtr1,rfr1,gbr1]
}
concl0 = pd.DataFrame(conc0, columns=['Holdout', 'CrossValidation'], index=['Linear','Lasso','Ridge','KNN','ADB', 'DT', 'RF','Gbr'])
```
